## From the value count tables it's clear that most attacks occur in the middle east and south asia. Most are bombings and armed assaults. Next I create a pivot table to look at how fatalities per attack type vary per region.  

```python
import numpy as np
pd.pivot_table(terrorism_df, values='nkill', index=['region_txt'], columns=['attacktype1_txt'], aggfunc=np.sum)
```

## The most deadly type of attack for most regions is from armed assault or bombing. South Asia has a high number of assassinations and the Middle East has a high number of kidnappings. Lastly, I'm interested in the success rate of each kind of attack. 

```python
terrorism_df.groupby('attacktype1_txt')['success'].mean()
```

## Hostage takings have the highest success rate and assassinations have the lowest. This seems to make sense because I would hypothesize assassination to be the most difficult to carry out.
